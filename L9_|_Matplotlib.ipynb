```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
fig, ax = plt.subplots(2,2)
print(ax.shape)
ax[0,0].plot(np.arange(6))          # top left
ax[0,1].plot(np.arange(6,0,-1))     # top right
ax[1,0].plot((-1)**np.arange(6))    # bottom left
ax[1,1].plot((-1)**np.arange(1,7))  # bottom right
plt.show()
```

```python
boysavg = [68, 67, 77, 61, 70]
div = ["A", "B", "C", "D", "E"]
avg = [70, 82, 73, 65, 68]
```

```python
index = np.arange(5)
width = 0.3
```

```python
plt.bar(index, avg, width, color = "red", label = "Divisions")
plt.bar(index+width, boysavg, width, color = "blue", label = "Boys Average")

plt.title("Bar Graph")
plt.xlabel("Divisions")
plt.ylabel("Average Marks")
plt.xticks(index + width/2, div)

plt.legend(loc = "best")
plt.show()
```

```python
import plotly.graph_objects as go

x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]

fig = go.Figure()

fig.add_trace(go.Scatter(x=x, y=y, mode='lines+markers', name='Line Plot'))

fig.update_layout(title='Dynamic Line Plot',
                  xaxis_title='X Axis',
                  yaxis_title='Y Axis')
fig.show()
```
